```python
import pandas as pd

testset = pd.DataFrame(
    {
        'sessionId' : ['100000000012', '100000000013', '100000000014', '100000000015', '100000000016'],
        'customerId' : ['woo23', 'woo23','woo23','nogame','nogame'],
        'inflowChannel' : ['Q2','Q6','Q2','B2','H0'],
        'date' : ['20201010', '20201010', '20201010', '20201014', '20210112']
     }
)

print(testset);
```

```python
testset = testset.sort_values(by = ['customerId','sessionId','date'])
```

```python
sequenceTable = testset.groupby(['customerId'])['inflowChannel'].apply(list).reset_index(name = 'sequence')
chk = pd.concat([
                    sequenceTable['customerId'],
                    pd.DataFrame(sequenceTable["sequence"].to_list(), columns=['seq_1','seq_2','seq_3'])
                ], axis = 1)
chk
```
